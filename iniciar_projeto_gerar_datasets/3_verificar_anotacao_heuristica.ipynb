```python
import json
from matplotlib import pyplot as plt
import cv2
import numpy as np
from os.path import exists
import os
import pandas as pd
import pdfplumber
```

```python
def open_json(path):
    with open(path, 'r') as openfile:
        json_object = json.load(openfile)
        return json_object
    
def wirte_json_file(jsd,path):
    class NpEncoder(json.JSONEncoder):
        def default(self, obj):
            if isinstance(obj, np.integer):
                return int(obj)
            if isinstance(obj, np.floating):
                return float(obj)
            if isinstance(obj, np.ndarray):
                return obj.tolist()
            return super(NpEncoder, self).default(obj)
    js = json.dumps(jsd, cls=NpEncoder,indent=4, ensure_ascii=False).encode('utf8')
    with open(path, "w") as outfile:
        outfile.write(js.decode())
def create_folder(path): 
    if not exists(path):
        os.makedirs(path)
    return path

def create_folder(path): 
    if not exists(path):
        os.makedirs(path)
    return path
```

```python
def get_page_img(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        image = pdf.pages[0].to_image()
        return image.original
```

```python
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
PURPLE = (25,12,80)
PINK = (255,6,231)
BROWN = (153,102,51)
YELLOW = (249,241,12)
GREY = (193,203,216)

HEADER = 'header'
FOOTER = 'footer'
TITLE = "title"
OTHER = "other"
EMENTA = "ementa"
ACORDAO = "acordao"
PARTES = "partes"
HEADING = "heading"

label_color_dict = {
HEADER :RED,
FOOTER : GREEN,
TITLE : BLUE,
OTHER : PINK,
EMENTA : PURPLE,
ACORDAO : BROWN,
PARTES :YELLOW ,
HEADING : GREY  
}
```

```python
create_folder('./an_heuri_imgs')
```

```python
an_heuri_path = 'heuristica_anotacao.json'
an_heuri = open_json(an_heuri_path)
```

```python
cols = ['id','anotated']
rows = []
for an in an_heuri:
    id = an['id']
    pdf_idtm_path = f'./ITD_MINI/{id}/{id}.pdf'
    an_path = f'./an_heuri_imgs/{id}.jpg'
    image = get_page_img(pdf_idtm_path)
    image = np.array(image)
    print(type(image))
    an_image = image.copy()
    ann = an['documento']
    thickness = 2
    for a in ann:
        rect = a['box']
        label = a['label']
        color = label_color_dict[label]
        sp = (rect[0],rect[1])
        ep = (rect[2],rect[3])
        an_image = cv2.rectangle(an_image, sp, ep, color, thickness)
    image_rgb = cv2.cvtColor(an_image,cv2.COLOR_BGR2RGB)
    cv2.imwrite(an_path, image_rgb)
    rows.append([id,an_path])
```

```python
def path_to_image_html(path):
    return '<img src="'+ path + '" width="600" >'
```

```python
df = pd.DataFrame(columns=cols,data=rows)
```

```python
doc_html_path = 'an_heuristica.html'
df.to_html(doc_html_path,escape=False, formatters={'anotated':path_to_image_html},index=False)
```
